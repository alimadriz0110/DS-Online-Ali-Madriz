```python
import pandas as pd
import lxml
import requests
from bs4 import BeautifulSoup
```

```python

url="https://www.as.com/"
respuesta=requests.get(url)

```

```python
def extraer_titulos(url):
    soup=BeautifulSoup(respuesta.text, "lxml")
    titulos=[]
    for heading in soup.find_all(['h1','h2', 'h3']):
        if heading.text.strip() and len(heading.text.strip())>15:
            titulos.append(heading.text.strip())

    for elemento in soup.select('.title, .headline, .article-title, .news-tittle'):
        if elemento.text.strip() and elemento.text.strip() not in titulos:
            titulos.append(elemento.text.strip())
    return titulos

    
```

```python
extraer_titulos(url)
```

```python
def extraer_articulos(url):
    soup=BeautifulSoup(respuesta.text, "lxml")
    articulos = []
    for articulo_elem in soup.select('article, .article, .post, .new-items, .a'):
        articulo = {}
        titulo_elem=articulo_elem.find(['h1','h2', 'h3']) or articulo_elem.select_one('.title, .headline')
        if titulo_elem:
            articulo['titulo']=titulo_elem.text.strip()
        else:
            continue
        articulos.append(articulo)

    return articulos
```

```python
extraer_articulos(url)
```
