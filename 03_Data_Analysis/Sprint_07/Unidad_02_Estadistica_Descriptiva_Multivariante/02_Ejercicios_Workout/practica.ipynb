```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data/tips.csv")
```

```python
from scipy import stats

def comprobar_supuestos(df, var_num, var_cat):
    """
    Devuelve normalidad por grupo (Shapiro) y homogeneidad de varianzas (Levene).
    """
    grupos = [g[var_num].values for _, g in df.groupby(var_cat)]
    nombres = list(df[var_cat].unique())

    print(f"--- Normalidad (Shapiro-Wilk) sobre '{var_num}' por '{var_cat}' ---")
    for nombre, g in zip(nombres, grupos):
        w, p = stats.shapiro(g)
        print(f"  {nombre}: W = {w:.4f}, p = {p:.4f}  -> {'normal' if p > 0.05 else 'NO normal'}")

    w, p = stats.levene(*grupos)
    print(f"\n--- Homogeneidad de varianzas (Levene) ---")
    print(f"  W = {w:.4f}, p = {p:.4f}  -> {'varianzas iguales' if p > 0.05 else 'varianzas DISTINTAS'}")

    return grupos
```
